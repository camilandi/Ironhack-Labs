```python
%%sql
SELECT au_lname, au_fname, pub_name, COUNT(title) as num_titles FROM authors
JOIN titleauthor
on titleauthor.au_id = authors.au_id
JOIN titles
on titleauthor.title_id = titles.title_id
JOIN publishers
ON titles.pub_id = publishers.pub_id
GROUP BY pub_name, au_lname, au_fname;
```

```python
%%sql
SELECT authors.au_id, au_lname, au_fname, SUM(qty) as number_of_sold_titles FROM authors
jOIN titleauthor
on titleauthor.au_id = authors.au_id
JOIN sales
ON titleauthor.title_id = sales.title_id
GROUP BY authors.au_id, au_lname, au_fname
ORDER BY number_of_sold_titles DESC
limit 5;
```

```python
%%sql
SELECT authors.au_id, au_lname, au_fname, SUM(qty) as number_of_sold_titles FROM authors
jOIN titleauthor
on titleauthor.au_id = authors.au_id
JOIN sales
ON titleauthor.title_id = sales.title_id
GROUP BY authors.au_id, au_lname, au_fname
ORDER BY number_of_sold_titles DESC;
```

```python
%%sql
SELECT authors.au_id, au_lname, au_fname, IFNULL(SUM(qty),0) as number_of_sold_titles FROM authors
RIGHT jOIN titleauthor
on titleauthor.au_id = authors.au_id
RIGHT JOIN sales
ON titleauthor.title_id = sales.title_id
GROUP BY authors.au_id, au_lname, au_fname

UNION

SELECT authors.au_id, au_lname, au_fname, IFNULL(SUM(qty),0) as number_of_sold_titles FROM authors
LEFT jOIN titleauthor
on titleauthor.au_id = authors.au_id
LEFT JOIN sales
ON titleauthor.title_id = sales.title_id
GROUP BY authors.au_id, au_lname, au_fname
ORDER BY number_of_sold_titles DESC;
```
