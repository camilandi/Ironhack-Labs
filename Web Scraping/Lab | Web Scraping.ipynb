```python
titles=[]
for i in soup.find_all('div',{'class':re.compile('usctitle')}):
    titles.append(i.text.strip())
```

```python
new_list=[]

for i in titles:
    new_list.append(re.findall('Title? \d\d?',i))
```

```python
soup_11 = init_bs(url_11)
soup_11
```

```python
tables = soup_11.find_all('table', {'class': 'wikitable'})[0]
tables
```

```python
rows = tables.find_all('tr')
rows_h = [i.text.strip().split('\n\n') for i in rows][0]
rows = [i.text.strip().split('\n\n') for i in rows][1:]
rows
```
