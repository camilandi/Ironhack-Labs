Which genre is most likely to contain free apps?
First, filter data where the price is 0.00. Assign the filtered data to a new variable called free_apps. Then count the values in free_apps

```python
free_apps = data[data['price'] == 0]
```

```python
free_apps
```

```python
len(free_apps.value_counts())
```

```python
free_apps['prime_genre'].value_counts()
```

Now you can calculate the proportion of the free apps in each genre based on the value counts you obtained in the previous two steps.

```python
proportion=(free_apps['prime_genre'].value_counts())/(data['prime_genre'].value_counts())
```

```python
proportion
```

If a developer tries to make money by developing and selling Apple Store apps, in which genre should s/he develop the apps? Please assume all apps cost the same amount of time and expense to develop.
We will leave this question to you. There are several ways to solve it.

```python

avg_price = data.groupby('prime_genre').mean().sort_values('price', ascending=False)['price'].reset_index()
avg_price.rename(columns={'price':'avg_price', 'prime_genre': 'genre'}).set_index('genre')
```
